```python
import numpy as np
import pandas as pd
import os

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 250)

pd.options.display.float_format = '{:.0f}'.format
```

```python
ncaa = pd.read_csv('~/Documents/Data Science/Data Projects/Capstone 1/data/CLEAN/ncaa_bios+stats_FINAL.csv', index_col=0)
ncaa.info()
ncaa.head()
```

```python
nfl = pd.read_csv('~/Documents/Data Science/Data Projects/Capstone 1/data/CLEAN/nfl_FINAL.csv', index_col=0)

nfl.info()
nfl.head()
```

```python
# nfl = nfl[nfl.ncaa_name.isin(ncaa.player)]


nfl = nfl.rename({'player':'nfl_player', 'height':'nfl_height', 'weight':'nfl_weight', 'position':'nfl_position'}, 
                axis='columns')
```

```python
nfl['in_nfl'] = True
nfl['success'] = [x > 2 for x in nfl.season_count]

cols = nfl.columns.tolist()
cols[1:1] = cols[-1:]
del cols[-1:]
nfl = nfl[cols]

len(nfl)
nfl.head()
```

```python
all_players = pd.merge(ncaa, nfl, how='left', left_on=['player', 'team'], right_on=['ncaa_name', 'college'],
                       validate='one_to_one')

# all_players = pd.merge(ncaa, nfl, how='left', left_on='player', right_on='ncaa_name', validate='one_to_one')
all_players.info()
all_players.head()
```

```python
no_stats = ncaa[ncaa.games.isnull()]
len(no_stats)
no_stats.head()

s = all_players.success.sum()
s
```
