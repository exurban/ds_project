## Merge NFL Data with NCAA Bios

```python
df = pd.merge(bios, nfl, how='left', left_on=['player', 'team'], right_on=['ncaa_name', 'college'],
                       validate='one_to_one')

df.info()
df.head()
```

## Feature Engineering

- drop unneeded cols
- standardize numeric columns
- one-hot encode home states
- one-hot encode teams
- one-hot encode positions

```python
# drop unneeded cols
df = df.drop(['player_id', 'player', 'class', 'home_town', 'home_country', 'last_school', 'first', 'last', 'stats_name', 
              'last_name', 'first_name', 'pref_name', 'nfl_player', 'nfl_position', 'nfl_height', 'nfl_weight', 
              'birth_date', 'birth_day', 'birth_month', 'birth_year', 'rookie', 'last_season', 'seasons', 'high_school', 
              'hs_state', 'season_count', 'in_nfl', 'college', 'ncaa_name'], axis = 1)
df.info()
```

```python


df.head(20)
```

```python
df = df.dropna(axis = 0)
df.info()
```

```python
pct_success = df.success.sum() / len(df.position)
pct_success
```

```python
from sklearn.preprocessing import LabelBinarizer

# create OH encoder for team
team_oh = LabelBinarizer()

# encode team
team_oh.fit_transform(df.team)

# create OH encoder for state
state_oh = LabelBinarizer()

# encode state
state_oh.fit_transform(df.home_state)

# create OH encoder for position
position_oh = LabelBinarizer()

# encode position
position_oh.fit_transform(df.position)
```
